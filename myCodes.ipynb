```python
recent_grads = recent_grads.dropna(axis=0) #remove rows with na data
cleaned_data_count = len(recent_grads) #count the number of rows in cleaned recent_grads
print(cleaned_data_count)
```

```python
import numpy as np
fig,ax = plt.subplots()
ind = np.arange(10)
width = 0.35
p1 = ax.bar(ind,recent_grads[:10]['Men'],width,color='blue')
p2 = ax.bar(ind + width,recent_grads[:10]['Women'],width,color="pink")
ax.set_xticks(ind+width)
ax.set_xticklabels(recent_grads[:10]['Major'],rotation=90)
plt.show()
```

```python
piv = recent_grads[:10].pivot_table(index='Major',values=['Men','Women','Rank'],aggfunc=np.sum)
piv = piv.sort_values(by=['Rank'])
piv.plot.bar(x=piv.index,y=['Men','Women'])
```

```python
table = recent_grads.pivot_table(index='Major_category',values=['Men','Women'],aggfunc=np.sum)
table.plot.bar(x=table.index,y=['Men','Women'],color=['blue','pink'])
```
